```python
import pandas as pd
df = pd.read_csv('product-info.csv')
df
```

```python
df['Total Sales'] = df['Price'] * df['Quantity Sold']
total_sales = df['Total Sales'].sum()
print('Total Sales:', total_sales)
```

```python
print('Average Sales Price:', df['Price'].mean())
```

```python
print('Standard Deviation of Sales Prices:', df['Price'].std())
```

```python
min_sales_price_row = df.loc[df['Total Sales'].idxmin()]
print('Product with the Minimum Sales Price:', min_sales_price_row['Product Name'])
```

```python
max_sales_price_row = df.loc[df['Total Sales'].idxmax()]
print('Product with the Maximum Sales Price:', max_sales_price_row['Product Name'])
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('house-info.csv')
df
```

```python
num = np.array(df['Price'])
num
```

```python
len(num)
```

```python
num1 = np.array([int(price.replace('$', '').replace(',', '')) for price in num])
num1
```

```python
np.percentile(num1,25), np.percentile(num1,50), np.percentile(num1,75)
```

```python
Q1 = np.percentile(num1,25)
Q3 = np.percentile(num1,75)

IQR = Q3 - Q1
upper_fence = Q3 + 1.5 * IQR
lower_fence = Q1 - 1.5 * IQR
```

```python
lower_fence, upper_fence
```

```python
num1[num1 > upper_fence]
```

```python
num1[num1<lower_fence]
```

```python
df = pd.read_csv('purchase-info.csv')
```

```python
df
```

```python
print('Average Purchase Amount for each category:')
print(df.groupby('Category')['Purchase Amount'].mean())
```

```python
category_counts = df['Category'].value_counts()
print('Most Frequent Purchase Category:', category_counts.idxmax())
print(df.groupby('Category').value_counts())
```

```python
most_frequent_category = df['Category'].mode()[0]
most_purchases_customer = df['Customer ID'].mode()[0]
category_counts = df['Category'].value_counts()
most_frequent_category = category_counts.idxmax()
customer_counts = df['Customer ID'].value_counts()
most_purchases_customer = customer_counts.idxmax()
print('Most Frequent Purchase Category:', most_frequent_category)
print('Customer with the Most Purchases:', most_purchases_customer)
```
